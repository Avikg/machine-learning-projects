```python
import numpy as np
import pandas as pd
```

```python
data = '/content/cross-validation.csv'
df = pd.read_csv(data)
df.head()
```

```python
df['Loan_Status'].replace({'Y':1, 'N':0}, inplace=True)
df['Gender'].replace({'Male':1, 'Female':0}, inplace=True)
df['Married'].replace({'Yes':1, 'No':0}, inplace=True)
df['Dependents'].replace({'3+':3}, inplace=True)
df['Education'].replace({'Graduate':1, 'Not Graduate':0}, inplace=True)
df['Self_Employed'].replace({'Yes':1, 'No':0}, inplace=True)
df['Property_Area'].replace({'Urban':2, 'Rural':1, 'Semiurban':0}, inplace=True)
df = df.drop('Loan_ID', axis=1)
```

```python
df = df.fillna(0)
```

```python
df.head()
```

```python
X = df.drop(columns=['Loan_Status']).values
y = df['Loan_Status'].values
print(X.shape)
print(y.shape)
```

```python
import warnings


# Ignore all warnings (not recommended)
warnings.filterwarnings("ignore")
```

```python
import numpy as np

# Step 1: Split the dataset into 80% for training and 20% for testing

# Set the random seed for reproducibility
np.random.seed(42)

# Shuffle the data
shuffled_indices = np.random.permutation(len(X))
X_shuffled = X[shuffled_indices]
y_shuffled = y[shuffled_indices]

# Split point for training and testing
split_point = int(0.8 * len(X_shuffled))

# Split the data
X_train, X_test = X_shuffled[:split_point], X_shuffled[split_point:]
y_train, y_test = y_shuffled[:split_point], y_shuffled[split_point:]

# Step 2: Train a Logistic Regression model with no regularization
from sklearn.linear_model import LogisticRegression

# Create the logistic regression model
model = LogisticRegression(solver='saga', penalty='none')

# Fit the model to the training data
model.fit(X_train, y_train)

# Step 3: Cross-validate the classifier with 5-folds
K = 5  # Number of folds
fold_size = len(X_train) // K

accuracies = []
precisions = []
recalls = []

for i in range(K):
    # Define the start and end indices for the current fold
    start = i * fold_size
    end = (i + 1) * fold_size

    # Create training and validation sets for this fold
    X_valid_fold = X_train[start:end]
    y_valid_fold = y_train[start:end]

    # Create training set for this fold (excluding the current fold)
    X_train_fold = np.concatenate([X_train[:start], X_train[end:]])
    y_train_fold = np.concatenate([y_train[:start], y_train[end:]])

    # Train the model on the training fold
    model.fit(X_train_fold, y_train_fold)

    # Predict on the validation fold
    y_pred_fold = model.predict(X_valid_fold)

    # Calculate accuracy, precision, and recall for class 1 (good)
    accuracy_fold = np.mean(y_pred_fold == y_valid_fold)
    precision_fold = np.sum((y_pred_fold == 1) & (y_valid_fold == 1)) / np.sum(y_pred_fold == 1)
    recall_fold = np.sum((y_pred_fold == 1) & (y_valid_fold == 1)) / np.sum(y_valid_fold == 1)

    accuracies.append(accuracy_fold)
    precisions.append(precision_fold)
    recalls.append(recall_fold)

# Calculate the mean accuracy, precision, and recall for class 1 (good)
mean_accuracy = np.mean(accuracies)
mean_precision = np.mean(precisions)
mean_recall = np.mean(recalls)

# Print the results
print(f"Mean Accuracy: {mean_accuracy:.2f}")
print(f"Mean Precision for Class 1 (good): {mean_precision:.2f}")
print(f"Mean Recall for Class 1 (good): {mean_recall:.2f}")

```
